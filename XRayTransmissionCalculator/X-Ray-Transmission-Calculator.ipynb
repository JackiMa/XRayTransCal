```python
from myinclude.mypkgs import * 
```

```python
# 读取数据

XrayMassCoef_csv_lists = r"./Data/XrayMassCoef_csv_lists/"
elements = ELEMENTS()
for element in os.listdir(XrayMassCoef_csv_lists):
    symbol, name, Z, A, _ = [par.strip() for par in element.split('_')]
    elements.add(ELEMENT(symbol, name, Z, A, pd.read_csv(XrayMassCoef_csv_lists+"/"+element)))
```

# 使用计算器

##  画元素的X射线质量衰减系数图像

```python
# 有三种获取元素的方式，分别是通过Z，符号symbol，名称name

hyd = elements.get(symbol='H')
hyd.draw()

car = elements.get(Z=6)
car.draw()

lead = elements.get(name='Lead')
lead.draw()
```

## Draw the transmittance of x-rays in a compound

```python
# config
chemical    = "Gd3Al2Ga3O12"
density     = 6.6 # g/cm3
thickness   = 1 # cm
Emin = 0.001 # MeV
Emax = 1 # MeV
precision = 1000 # points

# preparing
energy_lists = np.linspace(Emin, Emax, precision)
energy_lists = np.clip(energy_lists, Emin, Emax)  # The energy limit is Emin ~ Emax because the NIST data set is in this range

# draw
elements.draw_transmission(chemical, energy_lists, density, thickness)
```
